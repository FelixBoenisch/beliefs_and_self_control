Browsing through the STATA log files and also implementing the media regression in R, I found that the solution is not unique. Let's check for the range of coefficients that solve the minimization problem of the median regression.

```python
import numpy
constant = 15
result = []

for i in [x / 1000 for x in range(-2000, 6000, 1)]:
    estimate = constant + i*data_reg[(data_reg['exam2']==1) | (data_reg['exam2']==0)]['exam2']
    dev = abs(estimate-data_reg[(data_reg['exam2']==1) | (data_reg['exam2']==0)]['learneff_mikro5'])
    norm1 = numpy.linalg.norm(dev, ord=1)
    norm2 = numpy.linalg.norm(dev, ord=2)
    result.append([i, norm1, norm2])

res = pd.DataFrame(result, columns = ['coef', 'L1Norm', 'L2Norm'])

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(res['coef'], res['L1Norm'])
ax2.plot(res['coef'], res['L2Norm'])

min_L1 = res[res['L1Norm'] == res['L1Norm'].min()][['coef', 'L1Norm']]
min_L2 = res[res['L2Norm'] == res['L2Norm'].min()][['coef', 'L2Norm']]

for index, row in min_L1.iterrows():
    ax1.plot(row['coef'],row['L1Norm'],'ro', ms=0.5)
ax2.plot(min_L2['coef'], min_L2['L2Norm'],'ro')

ax1.title.set_text('Median Regression Solution')
ax2.title.set_text('OLS Regression Solution')

ax1.set(xlabel='coef', ylabel='L1-Norm')
ax2.set(xlabel='coef', ylabel='L2-Norm')
fig.tight_layout()
plt.show()

print('Coefficient range for L1-Norm Minimum: {} - {}'.format(min_L1['coef'].min(), min_L1['coef'].max()))
print('Coefficient for Unique L2-Norm Minimum: {}'.format(float(min_L2['coef'])))
```

This picture shows the multiplicity of solutions for the median regression. The regressions were made using  β0=15  as the constant. This yields the unique solution 3.583 for the OLS regression, which is only different from the coefficient in Table 1, Column 2 (coef = 3.633), because of the different constant (15 instead of 14.95). If we changed the constant to 14.95 or optimized over two parameters (for the L2-norm), the results would coincide. For the median regression with  β0=15 , the L1-Norm is minimized for  β1∈[0,4] . ## Tests

Statistical Test Recap: In the following I will be using tests on the mean and the distribution:

`Mann-Whitney U-Test (Wilcoxon Rank Sum test) - nonparametric` 
<a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mannwhitneyu.html#scipy.stats.mannwhitneyu">docu</a>

- **Null hypothesis:** Probability that a randomly drawn observation from one group is larger than a randomly drawn observation from the other is equal to 0.5 against an alternative that this probability is not 0.5. 
- **Assumptions:**  
    - Observations from both groups are independent of each other
    - Responses are ordinal
    - Under $H_0$ distributions are equal, under $H_1$ distributions are unequal
-  If the responses are assumed to be continuous and the alternative is restricted to a shift in location, then we can interpret the MWU-test to test for a difference in medians.

`Kruksal-Wallis-Test - nonparametric` 
<a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kruskal.html">docu</a>
- This test basically extends the MWU test to more than two groups of comparison. A significant KWT indicates that at least one sample stochastically dominates one other sample, but does not identify which one and how many. Hence, I will first use a Kruksal-Wallis-Test and then pairwise MWU tests.

`Two sample t-Test - parametric`
<a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html">docu</a>
- Tests for equality of means between two groups
- **Assumptions:**  
    - Samples should be drawn from the normal distribution (or normal distribution can be asumed via Central Limit Theorem)
    - Depending on the type of t-test, equal variance is assumed (if not, then use Welch test).
    - The two groups are independent

`One-way ANOVA - parametric`
<a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.f_oneway.html#scipy.stats.f_oneway">docu</a>
- This test extends the t-test to more than two groups of comparison. A significant ANOVA test indicates that at least one sample is drawn from a distribution with a different mean than another sample.
- **Assumptions:**  
    - The samples are independent.
    - Each sample is from a normally distributed population.
    - Homoscedasticity


`Two-sample Kolmogorov–Smirnov test - non-parametric`
<a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ks_2samp.html">docu</a>
- Tests for two samples coming from the same distribution, while being sensitive to location and shape of the distribution

MWU test is more outlier-robust since it is a rank-based test, but may have worse type I error control when data are both heteroscedastic and non-normal. We may want to consider as a robustness check to transform data to ranks and perform a t-test on the rank data.

## Ex-Post Rationalization

We start out by checking for correlation between belief inflation pre-exam and belief inflation post-exam (since beliefs go down it it mostly negative inflation = deflation), using all pre-exam periods and 1 post-exam period (grades not published yet in wave 4). Pre-exam inflation here is measures as return belief in pre-exam wave minus return belief in wave 2. Post-exam inflation is measured as before by return beliefs in the post-exam wave minus return beliefs in the pre-exam wave. 

Moreover, I create the variable *post_grade_infl*, which subtracts return beliefs pre-exam from those after knowing the grade (wave 5). Two important things:
1. I cannot obtain *post_grade_infl* for students in group 2, because at wave 6 grades were not published yet.
2. Students do not know their achieved points, just their grade. (What about Klausureinsicht?)

```python
#data_reg only includes main sample
data_reg['pre_exam_infl'] = np.nan
data_reg.loc[data_reg['group']==1, 'pre_exam_infl'] = data_reg[data_reg['group']==1]['exp_return3']-data_reg[data_reg['group']==1]['exp_return2']
data_reg.loc[data_reg['group']==2, 'pre_exam_infl'] = data_reg[data_reg['group']==2]['exp_return5']-data_reg[data_reg['group']==2]['exp_return2']

data_reg['post_exam_infl'] = np.nan
data_reg.loc[data_reg['group']==1, 'post_exam_infl'] = data_reg[data_reg['group']==1]['exp_return4']-data_reg[data_reg['group']==1]['exp_return3']
data_reg.loc[data_reg['group']==2, 'post_exam_infl'] = data_reg[data_reg['group']==2]['exp_return6']-data_reg[data_reg['group']==2]['exp_return5']

data_reg['post_grade_infl'] = np.nan
data_reg.loc[data_reg['group']==1, 'post_grade_infl'] = data_reg[data_reg['group']==1]['exp_return5']-data_reg[data_reg['group']==1]['exp_return3']
```

```python
data_reg['post_exam_effort'] = np.nan
data_reg.loc[data_reg['group']==1, 'post_exam_effort'] = data_reg[data_reg['group']==1]['expected_effort_mikro4']
data_reg.loc[data_reg['group']==2, 'post_exam_effort'] = data_reg[data_reg['group']==2]['expected_effort_mikro6']

list_of_reg_results = []
models = {
    'model1': ['points', ['pre_exam_infl']],
    'model2': ['points', ['post_exam_effort']],
    'model3': ['points', ['pre_exam_infl', 'overpred']],
    'model4': ['points', ['pre_exam_infl', 'post_exam_effort']],
    'model5': ['points', ['pre_exam_infl', 'overpred', 'geschlecht', 'age', 'semester', 'econ', 'business', 'first_mikro', 'd_mathe']]
}

for key in models.keys():
    est = sm.OLS(endog=data_reg[models[key][0]], exog=sm.add_constant(data_reg[models[key][1]])).fit(cov_type='HC1')
    est.summary()
    list_of_reg_results.append(est)

model_summary = Stargazer(list_of_reg_results)
model_summary.covariate_order(['const', 'pre_exam_infl', 'overpred', 'post_exam_effort', 'geschlecht', 'age', 'semester', 'econ', 'business', 'first_mikro', 'd_mathe'])
model_summary.show_degrees_of_freedom(False)
#model_summary.custom_columns([3, 1], [3, 1])

model_summary
```

Students that inflate their return beliefs more do not achieve higher points. Perhaps it is generally the worse students that suffer more from a self-control problem and distort beliefs to compensate for that. Students that overpredict their effort more, achieve lower points.
